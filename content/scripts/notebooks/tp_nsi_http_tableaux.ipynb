## Partie 1 – Analyser une requête GET

### 1.1 – Exemple guidé : extraire la méthode

Une requête HTTP brute ressemble à ceci (simplifiée) :

```
GET /search?q=python host=www.example.com path=/search
```

En Python, on peut la stocker dans une variable de type `str`.

La méthode `.split(' ')` découpe une chaîne en **tableau** (liste) de mots, en utilisant l'espace comme séparateur.

> 💡 `split()` renvoie une **liste** Python. On accède à chaque élément avec son **indice** entre crochets, en commençant à **0**.

Exécutez la cellule suivante pour observer le résultat :

````python
requete = "GET /search?q=python host=www.example.com path=/search"

elements = requete.split(' ')
print(elements)
# ['GET', '/search?q=python', 'host=www.example.com', 'path=/search']

methode = elements[0]   # premier élément, à l'indice 0
print(methode)          # GET
````

## Partie 2 – Analyser une requête POST et construire un dictionnaire

### 2.1 – Rappel : la méthode POST et le paramètre `maxsplit`

Contrairement à GET, la méthode **POST** envoie des données dans le **corps** de la requête (le *payload*). C'est utilisé notamment pour les formulaires de connexion, afin de ne pas faire apparaître le mot de passe dans l'URL.

Une requête POST simplifiée ressemble à :

```
POST /login host=www.lycee-pasteur.fr payload=identifiant=alice&motdepasse=1234
```

Remarquez que le payload contient lui-même des `=` : `payload=identifiant=alice&motdepasse=1234`. Si on fait simplement `.split('=')`, on obtiendrait trop de morceaux.

Pour éviter ça, `.split()` accepte un deuxième argument : le nombre **maximum de découpages** à effectuer. Avec `maxsplit=1`, on coupe uniquement au **premier** séparateur trouvé.

> 💡 `split(sep, maxsplit)` s'arrête après `maxsplit` coupures. Tout ce qui reste (y compris d'autres occurrences du séparateur) se retrouve dans le dernier élément.

Exécutez la cellule suivante pour comparer les deux comportements :

````python
s = "payload=identifiant=alice"

print(s.split('='))      # ['payload', 'identifiant', 'alice']  ← 3 morceaux
print(s.split('=', 1))   # ['payload', 'identifiant=alice']     ← 2 morceaux
````

## Partie 3 – Écrire une fonction robuste d'extraction

### 3.1 – Fonction fournie : `recherche_prefixe`

Jusqu'ici, on savait à l'avance à quel indice se trouvait le payload (indice 3). Mais dans une vraie requête, l'ordre des paramètres peut varier. On a donc besoin d'une fonction capable de **chercher** dans une liste l'élément qui commence par un mot donné.

La fonction suivante est **fournie** — vous n'avez pas à l'écrire, mais lisez-la attentivement avant de l'exécuter.

> 💡 `.startswith(prefixe)` est une méthode de chaîne qui renvoie `True` si la chaîne commence par `prefixe`, `False` sinon. Renvoyer `-1` quand on ne trouve pas est une convention classique en informatique.

````python
def recherche_prefixe(liste, prefixe):
    """
    Retourne l'indice du premier élément de `liste`
    qui commence par la chaîne `prefixe`.
    Retourne -1 si aucun élément ne correspond.
    """
    for i in range(len(liste)):
        if liste[i].startswith(prefixe):
            return i
    return -1
````

Voici des exemples d'utilisation à exécuter pour bien comprendre le comportement de la fonction :

````python
elements = ['POST', '/login', 'host=www.lycee-pasteur.fr', 'payload=identifiant=alice&motdepasse=1234']

i = recherche_prefixe(elements, 'payload')
print(i)              # 3
print(elements[i])    # payload=identifiant=alice&motdepasse=1234

i = recherche_prefixe(elements, 'host')
print(i)              # 2
print(elements[i])    # host=www.lycee-pasteur.fr

i = recherche_prefixe(elements, 'inconnu')
print(i)              # -1  (non trouvé)
````
