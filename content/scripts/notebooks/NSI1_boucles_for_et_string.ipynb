## d. Découpe d'une chaine: slice
La découpe d'une chaine texte se fait avec l'écriture:

```
texte[indice_debut : indice_fin + 1]
```

````python
texte1 = "je ne fais jamais"
texte2 = " mes devoirs"
texte3 = texte1 + texte2
print(texte3)
print(len(texte3))
````

````python
# pour extraire le premier mot, "je": 
# indice_debut = 0
# l'espace ' ' est le caractere d'index = 2
# donc le 'e' de 'je' est le caractere d'index = 1
# indice_fin + 1 = 1 + 1 = 2
texte3[0:2]
````

````python
# Le deuxieme mot:
# indice debut = 3
# indice fin = 5
texte3[ ?? ]
````

## e. découpe des mots d'une chaine
